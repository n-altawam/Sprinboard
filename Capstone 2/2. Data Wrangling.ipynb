## Data Cleaning



Ill start by inspecting the nulls.

```python
df.isnull().sum().sort_values(ascending=False)
```

The licence information is irrelevant to the project and will be dropped.

There are 13 entries under the `name` field that are missing, as well as 9393 entries related to reviews. I will insepct these to see whats missing.

```python
df = df.drop('license', axis=1)
```

```python
df[df.name.isna()]
```

There doesn't appear to be any visual correlation between these listings in terms of location or host. Most of them have no reviews, and without a description highlighting amenities, we wont be able to include them in the comparisons. For now, I will name them 'Unnamed Listing' and we'll decide if any value can be extracted during the EDA step.

```python
df.name.fillna('Unnamed Listing', inplace=True)
```

```python
df.price.describe()
```

```python
# The minimum value indicates there are listings for $0. 
df[df.price == 0]
```

It looks like there are 30 listings with a price of $0. This may be true or it may be an error, but there is no efficient way to check. Additionally, these entries are all listed as Hotel room for the `room_type`. Since the start-up will not be including these in their listing, we may drop them all together. For now, I will drop these entries since they wont help with the price.

```python
df = df[df.price != 0]
df.price.describe()
```

I will replace the null values in the `reviews_per_month` column with 0. As for the `last_review` column, it may be of help later in measuring the activity of listings but it will be dropped for now.

```python
df.reviews_per_month.fillna(0)
df = df.drop('last_review', axis=1)
df.columns
```

`id` and `name` are the only identifiers needed. The `host_name` and `host_id` will be dropped.

```python
df_final = df.drop(['host_name', 'host_id'], axis=1)
print(df_final.columns)
```

```python
df_detailed.columns()
```

The features I plan to investigate effect on price are: location, property type, and number of rooms.

For location: I'll be using the `neighbourhood` and `neighbourhood_group` as classifiers. The geocode location can be added later for a more in depth analysis or new feature engineering.

The `room_type` entry comprises of the following unique values: 'Entire home/apt', 'Private room', 'Hotel room', 'Shared room'. Hotel room entries will be dropped in a future step.

`accommodates` and `bedrooms` are integers relating to number of guests.
`bathrooms_text` contains string data about the number of baths.
These will also be added in the EDA step.

`amenities` is a json object of self-reported available amenities. This may be used in a future step.

To investigate local market conditions: I will investigate the hosts mentioned `availability_365`. This may be used to calculate occupancy rate in a future step.

Past rental data in this case will include: the `number_of_reviews` and `number_of_reviews_ltm` which is specific to the past 12 months.

The remaining features will be left in during the EDA step to check for other correlations. 

```python
listings_final = df_final.merge(df_detailed[['accommodates', 'bedrooms', 'bathrooms_text', 'amenities', 'id','beds']], on='id', how='left')
```
